```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import seaborn as sns

# Load the dataset
data = pd.read_csv("Recoded.csv")

# The dependant variable is continous and the independent variables are all categorical. 
categorical_vars = ['CustName', 'WHName', 'cityrecode', 'regionrecode', 'staterecode', 'productcoderecode']
continuous_var = 'QTY'

# Spliting into training and testing sets
seed = 42
train_data, test_data = train_test_split(data, test_size=0.2, random_state=seed)

# Preparing the data for training
X_train = train_data.drop(columns=[continuous_var])
y_train = train_data[continuous_var]
X_test = test_data.drop(columns=[continuous_var])
y_test = test_data[continuous_var]

# Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=seed)

# Fit the model to the training data
rf_model.fit(X_train, y_train)

# Predict demand using the Random Forest model
rf_predictions = rf_model.predict(X_test)

# Create a dataframe with actual and predicted demand
predictions = pd.DataFrame({'Actual_Demand': y_test, 'Predicted_Demand': rf_predictions})

# Visualize demand distribution using a histogram
plt.figure(figsize=(8, 6))
sns.histplot(data=predictions, x='Actual_Demand', bins=30, kde=True, color='blue', label='Actual Demand')
sns.histplot(data=predictions, x='Predicted_Demand', bins=30, kde=True, color='green', label='Predicted Demand')
plt.title("Demand Distribution (Actual vs. Predicted) - Random Forest")
plt.xlabel("Demand")
plt.ylabel("Frequency")
plt.legend()
plt.show()


```

```python
rmse_rf = mean_squared_error(y_test, rf_predictions, squared=False)
print("Random Forest RMSE:", rmse_rf)
```

```python
# Calculate the RMSE
rmse = mean_squared_error(predictions['Actual_Demand'], predictions['Predicted_Demand'], squared=False)
print(f"Root Mean Square Error (RMSE): {rmse}")
```
